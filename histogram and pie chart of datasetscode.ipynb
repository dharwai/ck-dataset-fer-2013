```python
import matplotlib.pyplot as plt
import os

# Defining paths and emotions
emotions = ["happy", "sad", "angry", "disgust", "neutral", "fear", "surprise"]
folder_path = "train/"

# Counting the number of images per emotion
counts = [len(os.listdir(folder_path + emotion)) for emotion in emotions]

# Plotting the bar chart
colors = ['red', 'yellow', 'black', 'blue', 'orange', 'green', 'pink']
plt.bar(emotions, height=counts, color=colors)
plt.ylabel('Number')
plt.xlabel('Emotions')

plt.savefig('hostgoarm.png')
plt.show()
```
